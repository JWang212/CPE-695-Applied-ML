## **2) Normal equation**

```python
import numpy as np
import matplotlib.pyplot as plt

x, y = np.loadtxt('C:\\Users\\acer\\Desktop\\HW2-F19 (1)\\data2.txt', delimiter=',', unpack=True)

plt.scatter(x,y,color='red',alpha='0.5')
x_bias = np.ones((len(x),1))
x = np.reshape(x,(len(x),1))
x = np.append(x_bias,x,axis=1)
x_transpose = np.transpose(x)
x_transpose_dot_x = x_transpose.dot(x)
temp_1 = np.linalg.inv(x_transpose_dot_x)
temp_2=x_transpose.dot(y)
theta =temp_1.dot(temp_2)
print("theta0 = {:0.5f} \ntheta1 = {:0.5f}".format(theta[0],theta[1]))

y = theta[0] + theta[1]*x
plt.plot(x,y,'b-')
plt.show()
```

Using normal equation, we found out that the coefficients are -3.895 & 1.193

## **3) Gradient Descent using BATCH MODE**

```python
import numpy as np
from sklearn.model_selection import train_test_split
import pandas as pd
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore", category=RuntimeWarning) 
```

```python
#loading data

X, y = np.loadtxt('C:\\Users\\acer\\Desktop\\HW2-F19 (1)\\data2.txt', delimiter=',', unpack=True)

X=X.reshape(-1,1)
y=y.reshape(-1,1)
```

```python
# split data into training and test data.

train_X, test_X, train_y, test_y = train_test_split(X, y, train_size=0.8, test_size=0.2,random_state=0)
```

```python
def cal_cost(theta,X,y):
    m = len(X)
    hypothesis = np.dot(X, theta)
    loss = hypothesis - y
    cost= np.sum(loss ** 2) / (2 * m)  # cost
    return cost
```

```python
def gradient_descent(X, y, theta, learning_rate, iterations):
    m = len(X)
    cost_history = np.zeros(iterations)
    theta_history = np.zeros((iterations, 2))
    for it in range(iterations):
        prediction = np.dot(X, theta)
        theta = theta - (1 / m) * learning_rate * (X.T.dot((prediction - y)))
        theta_history[it, :] = theta.T
        cost_history[it] = cal_cost(theta, X, y)
        #print ("iteration: %s | Cost function J: %.3f" % (it, cost_history[it])) 
    return theta, cost_history, theta_history
```

```python
#Training data | Printing iteration, cost function and coefficients

lr = 0.02
n_iter = 2000
theta = np.random.randn(2, 1)
X_b = np.c_[np.ones((len(train_X), 1)), train_X]
theta, cost_history, theta_history = gradient_descent(X_b, train_y, theta, lr, n_iter)
for it in range(n_iter):
    print ("iteration: %s | Cost function J: %.3f" % (it, cost_history[it]))
print("\n\nTheta0: {:0.3f},\nTheta1: {:0.3f}".format(theta[0][0], theta[1][0]))
print("Final cost function:  {:0.3f}".format(cost_history[-1]))
cost_history_train=cost_history
```

Answer: Here we see that the cost function becomes constant after almost 1000 iterations, so we can stop after 1000 iterations.

```python
#Testing data | Printing iteration, cost function and coefficients

lr = 0.02
n_iter = 2000
theta = np.random.randn(2, 1)
X_b = np.c_[np.ones((len(test_X), 1)), test_X]
theta, cost_history, theta_history = gradient_descent(X_b, test_y, theta, lr, n_iter)
for it in range(n_iter):
    print ("iteration: %s | Cost function J: %.3f" % (it, cost_history[it]))
print("\n\nTheta0: {:0.3f},\nTheta1: {:0.3f}".format(theta[0][0], theta[1][0]))
print("Final cost function:  {:0.3f}".format(cost_history[-1]))
cost_history_test=cost_history
```

## **3) Gradient Descent using STOCHASTIC MODE**

```python
def stochastic_gradient_descent(X, y, theta, learning_rate, iterations):
    m = len(X)
    cost_history = np.zeros(iterations)
    for it in range(iterations):
        cost = 0.0
        for i in range(m):
            rand_ind = np.random.randint(0, m)
            X_i = X[rand_ind, :].reshape(1, X.shape[1])
            y_i = y[rand_ind].reshape(1, 1)
            prediction = np.dot(X_i, theta)
            theta = theta - (1 / m) * learning_rate * (X_i.T.dot((prediction - y_i)))
            cost += cal_cost(theta, X_i, y_i)
        cost_history[it] = cost
        #print ("iteration: %s | Cost function J: %.3f" % (it, cost_history[it])) 
    return theta, cost_history
```

```python
#Training data | Printing Cost funtion, Iterations and coefficients

lr = 0.01
n_iter = 1000
theta = np.random.randn(2, 1)
X_b = np.c_[np.ones((len(train_X),1)),train_X]
theta,cost_history = stochastic_gradient_descent(X_b,train_y,theta,lr,n_iter)
for it in range(n_iter):
    print ("iteration: %s | Cost function J: %.3f" % (it, cost_history[it]))
print("Theta0: {:0.3f},\nTheta1:{:0.3f}".format(theta[0][0], theta[1][0]))
print("Final cost function:  {:0.3f}".format(cost_history[-1]))
cost_history_train_sto=cost_history
```

**a) Determine an appropriate termination condition (e.g., when cost function is less than a threshold, and/or after a given number of iterations).**

Answer: Here we see that the cost function becomes almost smooth after  300 iterations, so we can stop after 300 iterations.

**b) Print the cost function vs. iterations; compare and discuss the accuracy of batch and stochastic modes.**

Answer: The cost function vs iterations have been printed above. As compared to batch mode, stochastic mode is a little bit less accurate but it is almost there. In stochastic mode we can get our work done in less number of iterations, thus saving computing power.

```python
#Test data | Printing cost function, iterations and coefficients

lr = 0.01
n_iter = 1000
theta = np.random.randn(2, 1)
X_b = np.c_[np.ones((len(test_X),1)),test_X]
theta,cost_history = stochastic_gradient_descent(X_b,test_y,theta,lr,n_iter)
for it in range(n_iter):
    print ("iteration: %s | Cost function J: %.3f" % (it, cost_history[it]))
print("Theta0: {:0.3f},\nTheta1:{:0.3f}".format(theta[0][0], theta[1][0]))
print("Final cost function:  {:0.3f}".format(cost_history[-1]))
cost_history_test_sto=cost_history
```

```python
#Training and testing data plot

plt.plot(range(n_iter),cost_history_train_sto,'r-',label='training')
plt.plot(range(n_iter),cost_history_test_sto,'g-',label='test')
plt.xlabel('Iterations')
plt.ylabel('Cost Function J')
plt.title('Cost function J vs Iterations')
plt.legend()
#plt.ylim(0,3000)
plt.show()
```

Answer: Printing final cost function for different learning rates...

```python
lr_set = [0.01, 0.02, 0.04, 0.05, 0.06, 0.08]

for i in lr_set:
    lr = i
    n_iter = 1000
    cost_history_train_sto = 0
    theta = np.random.randn(2, 1)
    X_b = np.c_[np.ones((len(train_X),1)),train_X]
    theta,cost_history = stochastic_gradient_descent(X_b,train_y,theta,lr,n_iter)
    cost_history_train_sto=cost_history[-1]
    print("Cost function for lr = {} :  {:0.3f}".format(lr, cost_history_train_sto))
```
